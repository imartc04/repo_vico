```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.filters import unsharp_mask

img = cv2.imread('oveja.jpeg',0)

# global thresholding
kernel = np.ones((3, 3),np.uint8)

ret1,th1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
#th1 = cv2.dilate(th1, kernel,iterations = 1)

#th1 = unsharp_mask(th1, radius=1, amount=5)

th1 = cv2.erode(th1, kernel, iterations = 8)

l_segImg = img * th1

plt.imshow(l_segImg, "gray")

```
